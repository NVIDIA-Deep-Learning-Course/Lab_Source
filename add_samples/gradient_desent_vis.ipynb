업데이트할 w = Learning_Rate * (Y_Hat예측값 - Y실측값) * X)평균
업데이트할 b = Learning_Rate * (Y_Hat예측값 - Y실제값) * 1)평균 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 실제값 : 가설
X = np.random.rand(100)
Y = 0.2 * X + 0.5
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X, Y)
plt.show()
```

```python
def plot_prediction(pred, y):
    plt.figure(figsize=(8, 6))
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.scatter(X, y, label='Y : Hypothesis')
    plt.scatter(X, pred, label='Y_Hat : predict')
    # plt.plot(X, y, label='실제값')
    # plt.plot(X, pred, label='예측값')
    plt.legend()
    plt.show()
```

```python
## Gradient Descent
```

```python
w = np.random.uniform(-1, 1)
b = np.random.uniform(-1, 1)

learning_rate = 0.7

for epoch in range(200):
    Y_Hat = w * X + b
    
    error = np.abs(Y_Hat - Y).mean()
    if error < 0.001:
        break
        
    # gradient descent
    w_grad = learning_rate * ((Y_Hat - Y)*X).mean()
    b_grad = learning_rate * (Y_Hat - Y).mean()

    # W, b 갱신
    w = w - w_grad
    b = b - b_grad
    
    if epoch % 10 == 0:
        Y_Hat = w * X + b
        plot_prediction(Y_Hat, Y)
```
